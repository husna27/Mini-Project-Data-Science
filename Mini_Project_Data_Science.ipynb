# **4. Train the Model**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score

# Load dataset digits (gambar 8x8 dari angka 0-9)
digits = load_digits()
X, y = digits.data, digits.target

# Split data (70% training, 30% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Daftar model yang akan diuji
models = {
    "KNN": KNeighborsClassifier(n_neighbors=3),
    "SVM": SVC(kernel='rbf'),
    "Decision Tree": DecisionTreeClassifier(),
}

# Evaluasi model dengan cross-validation
results = {}
for name, model in models.items():
    scores = cross_val_score(model, X_train, y_train, cv=3)  # 5-fold cross-validation
    results[name] = np.mean(scores)  # Simpan rata-rata akurasi
    print(f"{name}: Mean Accuracy = {np.mean(scores):.4f}")

# Visualisasi hasil
plt.figure(figsize=(8, 5))
plt.bar(results.keys(), results.values(), color=['blue', 'green', 'red'])
plt.ylabel("Mean Accuracy")
plt.title("Comparison of Classification Models on Digits Dataset")
plt.ylim(0.8, 1.0)  # Rentang akurasi 80%-100%
plt.show()
```
